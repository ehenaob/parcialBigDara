```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
wiki_url = 'https://es.wikipedia.org/wiki/Anexo:Presidentes_de_Colombia'
table_class = 'wikitable'
```

```python
response = requests.get(wiki_url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
presidents_table = soup.find('table', attrs={'class': table_class})
print(presidents_table)
df = pd.read_html(str(presidents_table))
print(df)
```
